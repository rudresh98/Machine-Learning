```
import tensorflow as tf
```

```
#to assign fix value in tensor flow we need to define constant
x = tf.constant(30)
```

```
xl = tf.constant([3,5,66,88]) #muti value constant
```

```
y = tf.constant(20)
```

```
z = x+y #new tensor equation compile and store and wait for resources

#CPU,GPU
```

```
#we have another method to add the same
z1 = tf.add(x,y)
```

```
sess=tf.Session() #connect to cpu,gpu,tpu
```

```
#now we can our eqaution
output = sess.run(z)
```

```
output
```

```
#closing session
sess.close()
```

```
#we can perform same task in another way where we dont have to close the session
with tf.Session() as sess:
  output1=sess.run(z)
  tf.summary.FileWriter(output1)
```
